```python
# Importação das bibliotecas
import pandas as pd # Para a manipulação de DataFrames
import numpy as np # Cálculo algébricos
import matplotlib.pyplot as plt # Visualização dos dados
```

```python
# Criando um DataFrame do zero com um dicionário do Python

norte_brasil_1 = {
    'Siglas': ('AC','AP','AM','PA','RO','RR','TO'),
    'Estados': ('Acre','Amapá', 'Amazonas','Pará','Rondônia','Roraima','Tocantins'),
    'Capitais': ('Rio Branco','Macapá','Manaus','Belém','Porto Velho','Boa Vista','Palmas')
}

df1 = pd.DataFrame(norte_brasil_1)

df1
```

```python
norte_brasil_2 = {
    'Siglas': ('AC','AP','AM','PA','RO','RR','TO'),
    'Areas':(164123.040, 142828.521, 1559146.876, 1247954.666, 237590.547, 224300.506, 277720.520),
    'População':(869265, 829494, 4080611, 8513497, 1757589, 576568,1555229)
}

df2 = pd.DataFrame(norte_brasil_2)

df2
```

```python
# Juntando dois dataframes
norte_brasil = pd.merge(df1,df2)
norte_brasil
```

```python
gas = pd.read_csv('glp.csv', sep=';', encoding='latin1')
```

```python
gas
```

```python
gas.shape[0]
```

```python
gas.info()
```

```python
gas.columns
```

```python
gas.drop(['Região - Sigla','CNPJ da Revenda', 'Unidade de Medida', 'Bandeira'], axis=1, inplace = True)
```

```python
gas
```

```python
gas.drop(['Produto'], axis=1, inplace = True)
```

```python
gas
```

```python
gas.columns = ['Estado', 'Municipio', 'Revenda', 'Data','Venda', 'Compra']
```

```python
gas.dtypes
```

```python
gas['Venda'] = gas['Venda'].astype('float32')

gas['Compra'] = gas['Compra'].astype('float32')
```

```python
gas_am = gas[gas['Estado'] == 'AM']
```

```python
gas_am
```

```python
gas_am.describe()
```

```python
gas_am.isnull().sum()
```

```python
gas_am = gas_am.dropna(subset=['Compra'])
```

```python
gas_am['Lucro']=gas_am['Venda'] - gas_am['Compra']
```

```python
gas_am
```

```python
gas_am.sort_values(by='Venda', ascending=True)
```
